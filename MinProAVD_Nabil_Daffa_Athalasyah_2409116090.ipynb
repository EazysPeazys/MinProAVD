MENGUBAH TIPE DATA

```python
# Assuming your columns are named differently,
# you should replace the names in 'columns_to_convert' with the actual names in your DataFrame.
# For example, if your dataset has columns like 'Academic Reputation', 'Employer Reputation', etc.,
# your code should be modified like this:

columns_to_convert = ['Academic Reputation', 'Employer Reputation', 'Faculty Student', 'Citations per Faculty', 'International Faculty', 'International Students',  'Sustainability']

for col in columns_to_convert:
    # Handle missing values (if any) before converting to float
    df[col] = df[col].fillna(0).astype(str)  # Replace NaN with 0 and convert to string
    df[col] = df[col].str.replace('[\$,]', '', regex=True)
    df[col] = df[col].str.replace(r'\((.*?)\)', r'-\1', regex=True)
    df[col] = df[col].str.replace('-', '0')
    # Convert to float
    try:
        df[col] = df[col].astype(float)
    except ValueError:
        print(f"Warning: Could not convert column '{col}' to float. Skipping...")

print(df.dtypes)
```

```python
df.dtypes
```

```python
results = []

cols = df.select_dtypes(include=['float64'])

for col in cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = (len(outliers) / len(df)) * 100
    results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Create DataFrame after the loop has populated results
results_df = pd.DataFrame(results)
# Now 'Kolom' column exists, so you can set it as index
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

OUTLIERS

```python
columns_to_impute = ["Gross Sales", "Discounts", "Sales", "COGS", "Profit"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
df['Profit Margin (%)'] = (df['Profit'] / df['Sales']) * 100
df['Discount Rate (%)'] = (df['Discounts'] / df['Gross Sales']) * 100
df['COGS Percentage (%)'] = (df['COGS'] / df['Sales']) * 100

print(df.head())
```

```python
def categorize_discount(rate):
    if rate < 5:
        return 'Low Discount'
    elif 5 <= rate <= 15:
        return 'Medium Discount'
    else:
        return 'High Discount'

df['Discount Category'] = df['Discount Rate (%)'].apply(categorize_discount)
```
