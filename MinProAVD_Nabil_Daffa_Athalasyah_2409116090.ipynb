```python
import pandas as pd

# Membaca file CSV
file = "/content/2024 QS World University Rankings 1.1 (For qs.com).csv"
df = pd.read_csv(file)

# Membersihkan nama kolom dari spasi
df.columns = df.columns.str.strip()

# Mengecek apakah kolom yang ingin dihapus ada dalam DataFrame
columns_to_drop = ['2024 Rank', 'Overall Score']
existing_columns = [col for col in columns_to_drop if col in df.columns]

if existing_columns:
    df.drop(columns=existing_columns, axis=1, inplace=True)

# Menampilkan nama kolom setelah perubahan
print(df.columns)
```

```python
df.dtypes
```

MENGUBAH TIPE DATA

```python
columns_to_convert = [
    'Academic Reputation', 'Employer Reputation', 'Faculty Student',
    'Citations per Faculty', 'International Faculty',
    'International Students', 'Sustainability'
]

for col in columns_to_convert:
    if col in df.columns:
        # Mengganti NaN dengan 0 dan memastikan format string
        df[col] = df[col].astype(str).fillna('0')

        # Menghapus karakter yang bukan angka, titik, atau minus
        df[col] = df[col].str.replace(r'[^\d.-]', '', regex=True)

        # Mengonversi ke tipe float dengan error handling
        df[col] = pd.to_numeric(df[col], errors='coerce')
    else:
        print(f"Warning: Column '{col}' not found in DataFrame. Skipping...")

# Menampilkan tipe data setelah konversi
print(df.dtypes)
```

```python
df.dtypes
```

```python
import pandas as pd

# Menyimpan hasil analisis outlier
results = []

# Memilih hanya kolom bertipe numerik (float atau integer)
numeric_cols = df.select_dtypes(include=['number']).columns

for col in numeric_cols:
    # Menghitung Q1, Q3, dan IQR
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Menemukan outlier
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = (len(outliers) / len(df)) * 100

    results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Membuat DataFrame dari hasil analisis
results_df = pd.DataFrame(results)

# Menetapkan kolom 'Kolom' sebagai index
results_df.set_index('Kolom', inplace=True)

# Menampilkan hasil
display(results_df)

```

OUTLIERS

```python
columns_to_impute = [
    "Academic Reputation", "Employer Reputation", "Faculty Student",
    "Citations per Faculty", "International Faculty",
    "International Students", "Sustainability"
]

for col in columns_to_impute:
    if col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Menggunakan .loc[] untuk menghindari SettingWithCopyWarning
        df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    else:
        print(f"Warning: Column '{col}' not found in DataFrame. Skipping...")

# Menampilkan ringkasan data setelah capping
print(df.describe())
```

```python
results = []

# Pilih hanya kolom numerik (float64 dan int64)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Identifikasi outlier
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = (len(outliers) / len(df)) * 100

    results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Buat DataFrame dari hasil analisis
results_df = pd.DataFrame(results)

# Set kolom sebagai index
results_df.set_index('Kolom', inplace=True)

# Menampilkan hasil
display(results_df)
```
